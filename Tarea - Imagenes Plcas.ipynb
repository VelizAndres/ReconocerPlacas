```python
from matplotlib import pyplot as plt
from skimage import color
import cv2
import common
import pylab
import numpy as np
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'

```

```python
archivo = './auto1.jpg'
```

```python
auto_image = cv2.imread(archivo)
## plt.imshow(cv2.cvtColor(auto_image, cv2.COLOR_BGR2RGB));
auto_gris = cv2.cvtColor(auto_image, cv2.COLOR_BGR2GRAY)
plt.imshow(auto_gris, cmap = 'gray');
```

```python
th1 = 150
th2 = 200
#canny = cv2.Canny(auto_gris, 150, 200)
canny = cv2.Canny(auto_gris, th1, th2)
plt.imshow(canny, cmap = 'gray');
```

```python
#!pip install pytesseract
```

```python
canny = cv2.dilate(canny, None, iterations = 1)
plt.imshow(canny, cmap = 'gray');
```

```python
# _, cnts, _ = cv2.findContours(canny, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE) # OPENCV 3
cnts, _ = cv2.findContours(canny, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE) # OPENCV 4
```

```python
## Buscar pares, buscando el rectangulo de la placa
cnts
```

```python
for c in cnts:
    area = cv2.contourArea(c)
    x, y, w, h = cv2.boundingRect(c)
    epsilon = 0.09 * cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, epsilon, True)
    if len(approx) == 4 and area > 200:
        aspect_ratio = float(w)/h
        if (aspect_ratio > 1.0 and aspect_ratio < 2.0):
            print('area: ', area)
            print('w', w)
            print('h', h)
            print('x', x)
            print('y', y)
            print('Relacion de Aspecto:', aspect_ratio)
            cv2.drawContours(auto_image, [approx], 0, (0,255,0),3) 
            placa = auto_gris[y:y+h, x:x+w]
            text = pytesseract.image_to_string(placa, config = '--psm 11').strip()
            print('Placa: ', text)


        
```
